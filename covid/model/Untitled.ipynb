```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
icu_data = pd.read_csv("ICU(1).csv",index_col = False)
```

```python
icu_data.head()
```

```python
icu_data.drop("WINDOW",axis=1,inplace = True)
```

```python
icu_data.head()
```

```python
icu_data.isna().sum()
```

```python
icu_data.dropna(inplace=True)
```

```python
icu_data.drop_duplicates(keep='last',inplace=True)
```

```python
y = icu_data["ICU"]
x = icu_data.drop("ICU",axis=1)
```

```python
y
```

```python
x
```

```python
x.drop("PATIENT_VISIT_IDENTIFIER",axis=1,inplace=True)
```

```python
x
```

```python
x["AGE_PERCENTIL"] = x["AGE_PERCENTIL"].str.replace("th","")
x["AGE_PERCENTIL"] = x["AGE_PERCENTIL"].str.replace("Above 90","100")
```

```python
x["AGE_PERCENTIL"]
```

```python
x.dtypes
```

```python
x["AGE_PERCENTIL"] = x["AGE_PERCENTIL"].astype(str).astype(float).astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(42)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

## SVC

```python
from sklearn import svm
```

```python
clf = svm.SVC()
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf = GaussianNB()
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(random_state=0)
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
```

```python
np.random.seed(1)
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```
